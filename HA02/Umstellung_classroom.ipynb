# Umstellung der Hausaufgabenaus- und -abgabe auf https://classroom.gitlab.com

Beginnend mit Hausaufgabe 02 numpy wird die Ausgabe der Aufgabenstellung und die Abgabe der Ergebnisse über https://classroom.gitlab.com erfolgen. Die Einführung geschieht in den folgenden einfachen Schritten:

1. **Anlegen eines kostenlosen persönlichen Accounts bei https://github.com/**

    Sie können beliebige E-Mail-Adressen und beliebige Nutzernamen vergeben, weil Ihr github-Account mit einer classroom-Identität verknüpft wird

2. __Erzeugen eines Personal Access Tokens__ für den lesenden und schreibenden Zugriff aus Anwendungen heraus. Die offizielle Anleitung finden Sie auf [https://docs.github.com/](https://docs.github.com/de/authentication/keeping-your-account-and-data-secure/managing-your-personal-access-tokens). Wenn Sie mögen, können Sie bei der Gelegeheit auch die Sprache des Webinterfaces auf "Deutsch" einstellen. Eine vollständige bebilderte Anleitung finden Sie weiter unten in diesem Dokument

3. __"Annnehmen" der Hausaufgabe__

    durch Klicken auf diesen Link

    [https://classroom.github.com/a/OQXzkZR6](https://classroom.github.com/a/OQXzkZR6)

    Wenn Sie classroom zum ersten Mal besuchen, werden Sie aufgefordert, Ihren github-Account mit einer FH-E-Mail-Adresse (Export der Mitgliederdaten aus ILIAS) zu verknüpfen. Sollten Sie sich nicht finden, ist das nicht schlimm. Sie können auch so die Aufgabe annehmen. Weisen Sie mich dann bitte nur per Mail darauf hin, damit ich die Verknüpfung vornehmen kann.

4. __"Clonen" des Repos in Ihren `work/IT2_2324/`-Ordner__

    Sie erhalten einen github-Link auf Ihr individuelles `HA02` Repository. Navigieren Sie in Jupyterlab in Ihren `work/IT2_2324/`-Ordner und wählen Sie aus dem `git`-Menu die Option `Clone a Repository` (Bitte nicht `Download repository` ankreuzen). Hier müssen Sie erstmalig Ihren Personal Access Token benutzen.

5. __Bearbeiten der Aufgabe, Commit und Push__

    Bearbeiten Sie die Aufgabe bis zur Deadline (1.12.2023), commiten und pushen Sie. Sie dürfen beliebig viele Commits/ Pushs vornehmen. Geprüft wird der letzte, der vor der Deadline gepushed wird.

6. __Feedback abwarten, Verzeichnis löschen__

    Prüfen Sie von Zeit zu Zeit mit `Git` `pull` ob ein Feedback vorliegt (Hinweis: Für die HA02 gilt, dass Aufgabenteil a) für ein "bestanden" ausreicht). Danach ist das Repo hinfällig. (Keine weiteren push's mehr). Nach Abschluss des Kurses werden alle Hausaufgaben-Repos von github.com gelöscht. Wenn Sie die für sich aufbewahren wollen, kopieren Sie die Lösung am besten in Ihren Praktikum-Ordner und löschen das HA02 Verzeichnis komplett (Hinweis: Das geschieht über einen Konsolenbefehl `rm -rf ~/work/IT2_2324/[Ordnername]` (Zweimal prüfen, ob Sie das richtige Verzeichnis löschen!)




# Ausführliche Anleitung zum Erzeugen eines Personal Access Tokens in Github



1. __Klicken Sie auf https://github.com/ auf Ihr Avatar rechts oben im Bildrand__



2. __In dem sich öffnenden Profil-Menü klicken Sie auf `Settings`__



3. __Dann auf `Developer settings` (ganz unten)__



4. __Dann weiter navigieren auf `Personal access tokens` -> `Tokens (classic)`__



5. __Hier sehen Sie evtl. bereits eingerichtete Tokens. Mit `Generate new token (classic)` erzeugen Sie ein neues Token.__



    
Das, was Sie unter `Note` eintragen, wird anstelle des Benutzernamens, das erzeugte Token als Passwort bei einer Anmeldung aus einer Anwendung (auch Jupyterlab) heraus benutzt. Der Scope definiert die "Mächtigkeit" des neuen Zugangs. Die gezeigte Einstellung ist ausreichend.

6. __Vergessen Sie nicht, den Namen und das Token zu speichern__. 

    Das kann lokal geschehen oder auch als einfache Text-Datei in Ihrem `work`-Ordner
